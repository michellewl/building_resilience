# Commercial buildings: model interpretation

```python
from model_builder_new import *
import matplotlib
matplotlib.style.use('ggplot')
import xgboost
import shap
import pandas as pd
```

### Electric heating - SHAP feature explanation

```python
# Read in the data
predictors, targets, sc, names = get_targets_predictors('heating', 'elec')
```

```python
# Fit the optimal model 
X_train, X_test, y_train, y_test = train_test_split(predictors, targets, random_state=0)
model = xgboost.XGBRegressor(max_depth = 3, n_estimators = 250)
model.fit(X_train, y_train)
```

```python
# Explain the feature importance
shap_values = shap.TreeExplainer(model).shap_values(X_test)
shap.summary_plot(shap_values, X_test, feature_names = names, max_display = 5, plot_type = 'bar')
```
